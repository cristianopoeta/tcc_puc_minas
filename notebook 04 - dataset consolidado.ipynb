<hr style="height:0px;border:none;border-top:1px solid #c0c0c0;">

- Modelos de peças com `dim_z` zerada e respectiva correção.
- `dim_z == 0` pode ser corrigida com `dim_z = 1`, conforme https://www.bricklink.com/help.asp?helpID=261 :
> When the height of a piece is standard for how that piece is categorized, only 2 dimensions are used in the Item Name. For example, bricks that are 1 brick high, and plates that are 1 plate high have their height omitted from their Item Names. The first example above is described as a Brick 1 x 1, the height dimension being assumed as 1 and omitted.

```python
# identificação das peças a corrigir
sr_dim_z_0 = df_catal_bricks.dim_z == 0
df_catal_bricks[sr_dim_z_0]
```

```python
# correção e exibição dos valores corrigidos
df_catal_bricks.loc[df_catal_bricks.dim_z==0, 'dim_z'] = 1.0
df_catal_bricks[sr_dim_z_0]
```

<hr style="height:0px;border:none;border-top:1px solid #c0c0c0;">

- Modelos de peças com `weight` zerado e respectiva correção.

```python
# identificação das peças a corrigir
sr_weight_0 = df_catal_bricks.weight == 0
df_catal_bricks[sr_weight_0]
```

```python
# peças com mesmas dimensões que as peças com `weight` zerado,
# resgatando o endereço da imagem de thumbnail preservado em `df_url_img`
cols_merge = ['dim_x','dim_y','dim_z']
cols_index = cols_merge + ['weight']
df_similares = (
    df_catal_bricks
    .loc[sr_weight_0, cols_merge]
    .merge(df_catal_bricks.reset_index(), on=cols_merge)
    .iloc[:, :6]
    .merge(df_img_dscr.loc[:, ['thumbnail_url', 'description']], on='no')
    .assign(thumbnail_img=lambda df: '<img src="https:' + df.thumbnail_url + '"/>')
    .drop(columns='thumbnail_url')
    .set_index(cols_index)
    .sort_index()
)
table_styles = [
    {'selector': '.row_heading', 'props': 'vertical-align:top;border:1px solid black;'},
    {'selector': 'thead, .data', 'props': 'border:1px solid black;'},
    {'selector': 'thead th', 'props':'border-left:1px solid black;'},
    {'selector': '.col_heading, .data', 'props':'text-align:center;'}
]
df_similares.style.set_table_styles(table_styles)
```

```python
# valores para preenchimento de `weight == 0` com a mediana dos grupos de mesmas dimensões 
_ = slice(None)
df_fill_weight_0 = (
    df_similares.loc[(_, _, _, 0.0)]
    .reset_index(level='weight', drop=True)
).join(
    df_similares
    .reset_index()
    [lambda df: df.weight > 0]
    .groupby(cols_merge)
    [['weight']]
    .median()
).reset_index().set_index('no')
df_fill_weight_0.style
```

```python
# correção e exibição dos valores corrigidos
df_catal_bricks.loc[sr_weight_0, 'weight'] = df_fill_weight_0.weight
df_catal_bricks.loc[sr_weight_0] 
```

### Consolidação de `df_guia_precos` e `df_catal_bricks` em `df_vendas_bricks` 

- Junção das informações do histórico de vendas das peças de interesse com as características das peças de interesse constantes no catálogo de itens.

<hr style="height:0px;border:none;border-top:1px solid #c0c0c0;">

- Renomeação e reordenação das colunas para uma forma mais organizada e compacta.

```python
dct_rename ={
# variáveis da linha de pedido de peça (o = order)
'no': 'p_no',    # código do modelo da peça
'color_id': 'c_id',    # código/ID da cor da peça
'new_or_used': 'n_u',    # estado da peça
'ord_itm_id': 'o_itid',    # ID arbitrário da linha de pedido de peça
'quantity': 'o_qtty',    # quantidade
'unit_price': 'o_unpr',    # preço unitário
'seller_country_code': 'o_sctr',    # país do vendedor
'buyer_country_code': 'o_bctr',    # país do comprador
'date_ordered': 'o_dthr',    # data e hora do pedido
    
# características físicas da peça comercializada (p = part)
'weight': 'p_wt',    # peso/massa em gramas
'dim_x': 'p_dx',    # largura em studs
'dim_y': 'p_dy',    # comprimento em studs
'dim_z': 'p_dz',    # altura em bricks
    
# outros atributos da peça comercializada (p = part)
'name': 'p_nm',    # nome
'no_pfx': 'p_pfx',    # prefixo do código de modelo
'no_sfx': 'p_sfx',    # sufixo do código de modelo
    
# atributos da cor da peça comercializada (c = color)    
'color_name': 'c_nm',    # nome da cor
'color_code': 'c_cd',    # código da cor (RGB 24 bits hexadecimal)
'color_type': 'c_tp'    # tipo/família da cor
}
df_vendas_bricks = (
    df_vendas_bricks
    .rename(columns=dct_rename)
    [dct_rename.values()]
    .set_index('o_itid')
)
with pdoc('max_colwidth', None):
    h1s1t1(df_vendas_bricks)
```
